## 🧹 1.4. Limpieza y manejo de datos preliminares

Para la limpieza de datos realizaremos una pequeña imputación de los datos y llenaremos los datos que tengan valor nulo para mejorar los datos del dataset.

Nos enfocaremos en las columnas que usaremos en el analisis de gráficos para obtener mejores datos.

````python
# Nos enfocamos en la colmna Price para convertir su tipo de dato
def convert(x):
    # Aseguramos que sea string y limpiamos espacios o símbolos de moneda
    x = str(x).strip().replace('$', '').replace(',', '')

    if x == "0" or x == "":
        return 0.0

    # Manejo de sufijos (M para Millones, K para Miles)
    if 'M' in x:
        return float(x.replace('M', '')) * 1_000_000
    elif 'K' in x:
        return float(x.replace('K', '')) * 1_000

    try:
        return float(x)
    except ValueError:
        return None # O np.nan si prefieres marcar errores
# corregir algunos valores en formato string para que toda la columna sea tipo float
data.Price = data.Price.apply(convert)

# cast los datos de los Reviews a tipo entero
data.Reviews = data.Reviews.apply(convert)

# Convertir la columna installs a valores numericos
# ignoramos el +
def clean_installs(val):
    return int(val.replace(",","").replace("+","").replace("Free","0"))

data.Installs = data.Installs.apply(clean_installs)

# convertir las fechas en string a formato date
data['Last Updated'] = data['Last Updated'] = pd.to_datetime(data['Last Updated'], format='mixed', errors='coerce')

# Imprimimos las ultimas 5 columnas
data.describe()
````

## 🚀 1.5. Tratamiento de Valores Nulos

Identificar las columnas que presentan más del 30 % de valores nulos para decidir su eliminación.
Adicional, se reviso las columnas por la varianza para eliminar aquellas que no sean necesarias.
En el caso de columnas críticas con pocos valores faltantes, se realiza la imputación de datos.

````python
# Obtiene el número total de columnas del DataFrame original
total_columnas_antes = data.shape[1]
total_columnas_antes  # Muestra el total de columnas antes del tratamiento

# Calcula el porcentaje de valores nulos por cada columna
nulos = (
    data.isna()          # Identifica valores nulos (NaN)
      .mean()          # Calcula la proporción de nulos por columna
      .mul(100)        # Convierte la proporción a porcentaje
      .round(2)        # Redondea a 2 decimales
      .reset_index()   # Convierte el índice en una columna
)

# Renombra las columnas del DataFrame resultante
nulos.columns = ['Columna', 'Porcentaje_nulos']

# Selecciona las columnas que tienen más del 30% de valores nulos
columnas_mayor_30 = (
    nulos[nulos['Porcentaje_nulos'] > 30]['Columna']
    .tolist()          # Convierte los nombres de columnas a una lista
)

# Elimina del DataFrame las columnas con más del 30% de nulos
df_reducido = data.drop(columns=columnas_mayor_30)

# "Las columnas que tienen un valor único... probablemente sean inútiles"
for col in df_reducido.columns:
    if df_reducido[col].nunique() == 1:
        print(f"Eliminando columna de varianza cero: {col}")
        df_reducido = df_reducido.drop(columns=[col])

print(f"Dimensiones post-limpieza básica: {df_reducido.shape}")

# Obtiene el total de columnas después de eliminar las columnas con muchos nulos
total_columnas_despues = df_reducido.shape[1]
total_columnas_despues  # Muestra el total de columnas después del tratamiento

# Crea un DataFrame resumen con el antes y después del tratamiento de columnas
resumen_columnas = pd.DataFrame({
    'Estado': ['Antes del tratamiento', 'Después del tratamiento'],
    'Total_columnas': [total_columnas_antes, total_columnas_despues]
})

# Muestra el resumen final
resumen_columnas
````

Luego de validar, nos percatamos que no existen columnas para eliminar.

### Filtrado por Umbral de Nulos (30%)
Se decidió eliminar automáticamente cualquier columna que presentara más del 30% de valores ausentes.a
###¿Por qué?:
Una columna con un alto índice de nulos introduce sesgos y reduce la confiabilidad de cualquier modelo o gráfico. Al superar el 30%, la cantidad de datos "inventados" o imputados necesarios para rellenar.
### Reducción de Dimensionalidad
El paso de total_columnas_antes a total_columnas_despues no es solo una limpieza, es una optimización.
### ¿Por qué?:
Trabajar con un DataFrame más ligero permite que las operaciones en Python sean más rápidas y que el enfoque del análisis se mantenga en las variables que realmente tienen información completa y útil.

### 🚀 1.9.3. **Pregunta 3.** Para fines del componente práctico, consideraremos que todos los datos por debajo o encima de los límites inferiores o superiores del rango intercuartílico serán considerados outliers. Elimina estos datos. ¿Cúantos registros eran? (Considera solo la variable `Price`)

Nota: Supondremos un rango intercuartílico leve, donde el extremo inferior está dado por:
$$Q_1-1.5*IQR$$
y el extremo superior por:
$$Q_3 + 1.5*IQR$$

donde $$IQR=Q_3-Q_1$$



````python
# Calcula e imprime los valores inferiores y superiores del rango intercuartilico
# para las siguientes variables: Price
# No olvides que solo consideraremos las aplicaciones pagas (excluir las gratuitas)
#### TAREA 3 ###############
# TU CÓDIGO AQUÍ

# Obtenemos las apps que son pagadas
data_free_apps = data[data.Type == 'Paid']

#Se reutiliza el metodo de deteccion de outliers revisado en clase de arquitectura de datos
def detectar_outliers_iqr(dataframe, columna):
    Q1 = dataframe[columna].quantile(0.25)
    Q3 = dataframe[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = dataframe[(dataframe[columna] < limite_inferior) | (dataframe[columna] > limite_superior)]
    return outliers, limite_inferior, limite_superior

# Visualización antes de tratar
plt.figure(figsize=(10, 5))
sns.boxplot(x=data_free_apps['Price'])
plt.title("Boxplot Price (con Outliers)")
plt.show()

# Detectar
outliers_ingreso, min_value, max_value = detectar_outliers_iqr(data_free_apps, 'Price')
print(f"Outliers detectados en Price: {len(outliers_ingreso)}")
print(f"Inferior: {min_value}")
print(f"Superior: {max_value}")

#############################
````

````python
# Tratamiento: Clipping (tope) en lugar de eliminar
data_free_apps['Price'] = data_free_apps['Price'].clip(lower=min_value, upper=max_value)
print(f"Existen {data_free_apps.shape[0]} registros luego del filtro y eliminar los outliers")

# Visualización post-tratamiento
plt.figure(figsize=(10, 5))
sns.boxplot(x=data_free_apps['Price'])
plt.title("Boxplot Price (Corregido)")
plt.show()
````

#### 🧠 Interpretación

Los dos boxplots muestran claramente el impacto de los valores atípicos en la variable Price. En el primer gráfico (Boxplot Price con Outliers) se observa una fuerte asimetría hacia la derecha, donde la mayoría de los valores se concentran en un rango bajo de precios, mientras que existen múltiples outliers muy alejados que alcanzan valores cercanos a 400. Esto provoca que la caja y los bigotes queden comprimidos cerca del origen, dificultando visualizar adecuadamente la distribución central de los datos. Además, el cálculo basado en el IQR (Interquartile Range) detectó 89 valores atípicos, con límites aproximados de -3.40 y 10.03, indicando que cualquier precio superior a ese umbral se considera extremo respecto a la distribución principal.

En el segundo gráfico (Boxplot Price Corregido), tras eliminar esos outliers, la distribución se vuelve mucho más interpretable, permitiendo observar con claridad el rango intercuartílico y la posición de la mediana. La caja ahora refleja mejor la dispersión real de los datos y los bigotes se extienden dentro de un rango más representativo del comportamiento típico del precio. En total quedan 794 registros, lo que sugiere que la mayoría de los datos se concentran en valores relativamente bajos y que los precios extremadamente altos correspondían a casos aislados que distorsionaban el análisis inicial. Este proceso mejora la calidad del análisis exploratorio y la estabilidad de futuros modelos de machine learning.

**Análisis complementario con Heatmap**

>Una vez identificadas visualmente las posibles relaciones en el pairplot, se utilizó un heatmap de correlación para cuantificar con precisión la fuerza y dirección de esas relaciones. En contraste, el Rating demostró no tener correlación significativa con ninguna otra variable, siendo el valor más alto apenas 0.15 con Reviews, lo que confirma que la calidad percibida por los usuarios no depende del precio, tamaño, ni la popularidad de la app.

>Un hallazgo interesante fue la correlación negativa entre Price e Installs de -0.15, que aunque débil debido al rango en que se encuentra su valor, indica que el precio es una barrera que iniside en la selección de los usuarios y que las apps gratuitas tienden a tener más descargas. **Pregunta 7:** Dónde hay más competencia?, que porcentajes?

>El mercado de Google Play está altamente concentrado en pocas categorías. La categoría FAMILY es la más competida con 1,168 apps que representan el 19.5% del dataset, lo que la convierte en el mercado más saturado y difícil de ingresar.

````python
# Genera aquí el código para gráficar la Proporción de las categorias de las
# aplicaciones en la Google Store

##### TAREA 11 #####################
# TU CÓDIGO AQUÍ

# Contar las apps por categoria
category_counts = data.Category.value_counts()

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Grafico 1
sns.barplot(x=category_counts.values, y=category_counts.index, ax=axes[0], palette='viridis')
total = category_counts.sum()

axes[0].set_title('Número de Apps por Categoría', fontsize=14, pad=15)
axes[0].set_xlabel('Cantidad de apps', fontsize=12)
axes[0].set_ylabel('')
axes[0].tick_params(axis='y', labelsize=9)

# Agregar valores al final de cada barra
for idx, valor in enumerate(category_counts.values):
    axes[0].text(valor + 5, idx, f'{valor} ({valor/total*100:.1f}%)',
                va='center', fontsize=8)

# Gráfico 2 — Top 10 pie chart mejorado
top10 = category_counts.head(10).copy()
otros = category_counts[10:].sum()
top10['OTROS'] = otros

# Explotar la categoría más grande para destacarla
explode = [0.05] * len(top10)

wedges, texts, autotexts = axes[1].pie(
    top10.values,
    labels=None,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    pctdistance=0.85 # ubicacion %
)

# Ajustar tamaño de porcentajes
for autotext in autotexts:
    autotext.set_fontsize(11)

# Leyenda separada para etiquetas
axes[1].legend(wedges, top10.index,
               title='Categorías',
               loc='center left',
               bbox_to_anchor=(-0.4, 0.5), #ubicacion legend
               fontsize=10) #tamaño de letar legend
axes[1].set_title('Proporción Top 10 Categorías', fontsize=14, pad=15)

plt.tight_layout()
plt.show()

# Imprimir top 10 con porcentajes
total = category_counts.sum()
print("\nTop 10 categorías más competidas:")
for cat, count in category_counts.head(10).items():
    print(f"{cat:<25} {count:>5} apps  ({count/total*100:.1f}%)")

#############################
````

````python
# Genera aquí el código para gráficar la relación de cantidades con las categorías
# tanto de número total de instalaciones y precio promedio por categoría.

# en este punto para relacionar más fácilmente la información de las dos partes
# construir una figura en matplotlib con dos filas y una columna. En el lienzo 1
# grafica el número total de instalaciones por categoría y en el lienzo 2
# grafica el precio promedio por categoría

# hint: Para crear el gráfico de dos filas y una columna usen plt.subplots
# y luego envien cada lienzo como parámetro a la grafica de seaborn de esta manera:
# sns.graficoausar(....., ax=axes[0])

# Usen pandas para generar el conteo total de número de instalaciones y
# promedio por categorías usando la función groupby despues gráfica esta
# información con seaborn

##### TAREA 12 #####################
# TU CÓDIGO AQUÍ

# Metricas por categoria
installs_by_category = data.groupby('Category')['Installs'].sum().sort_values(ascending=False)
price_by_category = data.groupby('Category')['Price'].median().sort_values(ascending=False)

fig, axes = plt.subplots(2, 1, figsize=(10, 14))

# Gráfico 1 — Total instalaciones (escala en millones)
installs_millions = installs_by_category / 1_000_000

#Creación de barplot
sns.barplot(x=installs_millions.values, y=installs_millions.index, ax=axes[0], palette='Blues_r')

axes[0].set_title('Total de Instalaciones por Categoría', fontsize=14, pad=15)
axes[0].set_xlabel('Total Installs (Millones)', fontsize=12)
axes[0].set_ylabel('')
axes[0].tick_params(axis='y', labelsize=9)

for idx, valor in enumerate(installs_millions.values):
    axes[0].text(valor + 0.5, idx, f'{valor:.1f}M', va='center', fontsize=8)

# Gráfico 2 — Precio por categoría usando la mediana
# Precio promedio solo apps pagas por categoría
paid_data = data[data.Type == 'Paid']
price_by_category = paid_data.groupby('Category')['Price'].mean().sort_values(ascending=False)

# Creacion Barplot
sns.barplot(x=price_by_category.values, y=price_by_category.index, ax=axes[1], palette='Oranges_r')

axes[1].set_title('Precio Promedio por Categoría (solo apps pagas)', fontsize=14, pad=15)
axes[1].set_xlabel('Precio Promedio (USD)', fontsize=12)
axes[1].set_ylabel('')
axes[1].tick_params(axis='y', labelsize=9)

for idx, valor in enumerate(price_by_category.values):
    axes[1].text(valor + 0.01, idx, f'${valor:.2f}', va='center', fontsize=8)

plt.tight_layout()
plt.show()

#############################
````

>La categoría GAME presenta en promedio el mayor tamaño de aplicación. Esto se observa porque:

>*   La mediana del tamaño en GAME es claramente mayor que en MEDICAL y PRODUCTIVITY.
>*   La densidad principal de GAME se concentra aproximadamente entre 20,000 KB y 40,000 KB.
>*   Además, GAME muestra una mayor presencia de aplicaciones grandes, alcanzando valores cercanos a 50,000 KB.
Esto es coherente con la naturaleza de cada categoría, ya que los juegos suelen requerir más recursos gráficos, animaciones y archivos multimedia, lo que incrementa el tamaño de la aplicación.

>En cambio:

>*  MEDICAL presenta tamaños más moderados, concentrados entre 5,000 KB y 25,000 KB.
>*  PRODUCTIVITY muestra principalmente aplicaciones pequeñas, con una gran concentración por debajo de 10,000 KB.

>Con respecto a la pregunta **¿Cuál categoría presenta un sector más exigente?**

>La categoría MEDICAL parece representar el sector más exigente en términos de calificaciones (ratings). En el gráfico de Distribución de Rating por Categoría se observa que:

>*  La mediana del rating en MEDICAL es ligeramente más alta, cercana a 4.4.
>*  La distribución está más concentrada en valores altos (entre 4.2 y 4.6).
>*  Hay menos dispersión hacia ratings bajos en comparación con otras categorías.

>En contraste:

>GAME presenta mayor variabilidad en las calificaciones, incluyendo más valores por debajo de 4.0, lo que indica que la calidad de los juegos puede ser más heterogénea.

>PRODUCTIVITY también muestra algunos ratings bajos, lo que sugiere una mayor variabilidad en la satisfacción de los usuarios.

````python
# Genera aquí el código para comparar las distribuciones del tamaño en las 3
# categorias de interés: 'PRODUCTIVITY', 'MEDICAL', 'GAME'

##### TAREA 13 #####################
# TU CÓDIGO AQUÍ
# Filtrar solo las categorías de interés
categories_of_interest = ['PRODUCTIVITY', 'MEDICAL', 'GAME']
data_filtered = data[data['Category'].isin(categories_of_interest)]

# Distribución de Size por categoría
fig, ax = plt.subplots(figsize=(14, 6))

sns.violinplot(x='Category', y='Size', data=data_filtered, ax=ax, palette='Blues', inner='box', cut=0)

ax.set_title('Distribución de Size por Categoría', fontsize=14, pad=15)
ax.set_xlabel('Categoría', fontsize=12)
ax.set_ylabel('Size (KB)', fontsize=12)

plt.tight_layout()
plt.show()

# Imprimir estadísticas
print("\nEstadísticas de Size por categoría:")
print(data_filtered.groupby('Category')['Size'].describe().round(2))

#############################
````

````python
# Genera aquí el código para comparar las distribuciones del rating en las 3
# categorias de interés: 'PRODUCTIVITY', 'MEDICAL', 'GAME'

##### TAREA 14 #####################
# TU CÓDIGO AQUÍ
# Distribución de Rating por categoría
categories_of_interest = ['PRODUCTIVITY', 'MEDICAL', 'GAME']
data_filtered = data[data['Category'].isin(categories_of_interest)]

fig, ax = plt.subplots(figsize=(14, 6))

sns.violinplot(x='Category', y='Rating', data=data_filtered, ax=ax, palette='Oranges', inner='box', cut=0)

ax.set_title('Distribución de Rating por Categoría', fontsize=14, pad=15)
ax.set_xlabel('Categoría', fontsize=12)
ax.set_ylabel('Rating', fontsize=12)

plt.tight_layout()
plt.show()

# Estadísticas de Rating por categoría
print("\nEstadísticas de Rating por categoría:")
print(data_filtered.groupby('Category')['Rating'].describe().round(2))

#############################
````
